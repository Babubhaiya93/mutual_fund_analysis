```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/performance_metrics.csv")

df.head()
```

```python
df_sorted = df.sort_values(
    by='total_return_%',
    ascending=False
)

df_sorted
```

```python
plt.figure(figsize=(10,5))

plt.bar(
    df_sorted['fund_name'],
    df_sorted['total_return_%']
)

plt.xticks(rotation=45)
plt.title("Total Return Comparison")
plt.ylabel("Return (%)")
plt.tight_layout()
plt.show()
```

```python
df['growth_multiple'] = (
    df['end_nav'] /
    df['start_nav']
)

df[['fund_name','growth_multiple']]
```

```python
plt.figure(figsize=(10,5))

plt.bar(
    df['fund_name'],
    df['growth_multiple']
)

plt.xticks(rotation=45)
plt.title("NAV Growth Multiple")
plt.ylabel("Growth Factor")
plt.tight_layout()
plt.show()
```

```python
corr = df[
    ['start_nav','end_nav','total_return_%']
].corr()

corr
```

```python
plt.figure(figsize=(6,4))

sns.heatmap(
    corr,
    annot=True
)

plt.title("Correlation Matrix")
plt.show()
```

```python
df['rank'] = df[
    'total_return_%'
].rank(
    ascending=False
)

df.sort_values('rank')
```

```python
best_fund = df.loc[
    df['total_return_%'].idxmax()
]

print(best_fund)
```

```python
df.to_csv(
    "../reports/advanced_analytics_results.csv",
    index=False
)

print("Advanced Analytics Saved")
```
