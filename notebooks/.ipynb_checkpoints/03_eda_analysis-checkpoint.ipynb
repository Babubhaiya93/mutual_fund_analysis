```python
import pandas as pd
import sqlite3

conn = sqlite3.connect("../data/db/bluestock_mf.db")

df = pd.read_sql(
    "SELECT * FROM nav_history",
    conn
)

df.head()
```

```python
print("Shape:", df.shape)

print("\nColumns:")
print(df.columns)

print("\nData Types:")
print(df.dtypes)

print("\nMissing Values:")
print(df.isnull().sum())
```

```python
df["date"] = pd.to_datetime(df["date"], dayfirst=True)

df = df.sort_values("date")

df.head()
```

```python
df["daily_return"] = (
    df.groupby("fund_name")["nav"]
      .pct_change()
)

df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

for fund in df["fund_name"].unique():

    temp = df[df["fund_name"] == fund]

    plt.figure(figsize=(10,5))

    plt.plot(temp["date"], temp["nav"])

    plt.title(f"{fund} NAV Trend")

    plt.xlabel("Date")
    plt.ylabel("NAV")

    plt.show()
```

```python
df.to_csv(
    "../data/processed/clean_nav_data.csv",
    index=False
)

print("Clean data saved successfully")
```

```python
import pandas as pd
import plotly.express as px

df = pd.read_csv("../data/processed/clean_nav_data.csv")

df["date"] = pd.to_datetime(df["date"], format="mixed")

fig = px.line(
    df,
    x="date",
    y="nav",
    color="fund_name",
    title="Daily NAV Trend of Mutual Funds"
)

fig.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))

sns.histplot(
    df["daily_return"],
    bins=50,
    kde=True
)

plt.title("Distribution of Daily Returns")
plt.show()
```

```python
plt.figure(figsize=(12,6))

sns.boxplot(
    data=df,
    x="fund_name",
    y="nav"
)

plt.xticks(rotation=45)

plt.title("NAV Distribution by Fund")

plt.show()
```

```python
volatility = df.groupby(
    "fund_name"
)["daily_return"].std()

volatility.sort_values().plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Volatility Comparison")

plt.show()
```

```python
returns = df.pivot_table(
    index="date",
    columns="fund_name",
    values="daily_return"
)

corr = returns.corr()

plt.figure(figsize=(8,6))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm"
)

plt.title("Fund Return Correlation Matrix")

plt.show()
```

```python
avg_nav = df.groupby(
    "fund_name"
)["nav"].mean()

avg_nav.sort_values(
    ascending=False
).plot(
    kind="bar",
    figsize=(10,5)
)

plt.title("Average NAV by Fund")

plt.show()
```

```python
df["month"] = df["date"].dt.to_period("M")

monthly = df.groupby(
    ["month","fund_name"]
)["nav"].mean().reset_index()

monthly["month"] = monthly["month"].astype(str)

fig = px.line(
    monthly,
    x="month",
    y="nav",
    color="fund_name",
    title="Monthly NAV Trend"
)

fig.show()
```

```python
plt.savefig("../reports/nav_trend.png", dpi=300, bbox_inches="tight")
plt.show()
```
