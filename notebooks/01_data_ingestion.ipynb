```python
import pandas as pd
import glob
```

```python
files = glob.glob("../data/raw/*.csv")

print("Files found:")
for f in files:
    print(f)
```

```python
df_list = []

for file in files:
    temp = pd.read_csv(file)
    temp["fund_name"] = file.split("\\")[-1].replace(".csv", "")
    df_list.append(temp)

df = pd.concat(df_list, ignore_index=True)

df.head()
```

```python
print("Shape:", df.shape)
print(df.info())
```

```python
df.to_csv("../data/processed/combined_nav_data.csv", index=False)

print("Data saved successfully")
```
